```python


#      3  Is there a relationship between "Confidence_Score" and the correctness of the label (e.g., higher confidence for "Correct" labels)?




import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
df = pd.read_csv('ai_computer_vision_dataset.csv')
analysis_df = df[['Label', 'Confidence_Score']].dropna()
correct_stats = analysis_df[analysis_df['Label'] == 'Correct']['Confidence_Score'].describe()
incorrect_stats = analysis_df[analysis_df['Label'] == 'Incorrect']['Confidence_Score'].describe()
uncertain_stats = analysis_df[analysis_df['Label'] == 'Uncertain']['Confidence_Score'].describe()
print("Confidence Score Statistics:")
print("\nCorrect labels:")
print(correct_stats)
print("\nIncorrect labels:")
print(incorrect_stats)
print("\nUncertain labels:")
print(uncertain_stats)


```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Label', y='Confidence_Score', data=analysis_df, order=['Correct', 'Incorrect', 'Uncertain'])
plt.title('Confidence Score Distribution by Label Correctness')
plt.show()
correct_scores = analysis_df[analysis_df['Label'] == 'Correct']['Confidence_Score']
incorrect_scores = analysis_df[analysis_df['Label'] == 'Incorrect']['Confidence_Score']
uncertain_scores = analysis_df[analysis_df['Label'] == 'Uncertain']['Confidence_Score']
f_val, p_val = stats.f_oneway(correct_scores, incorrect_scores, uncertain_scores)
if p_val < 0.05:
    from statsmodels.stats.multicomp import pairwise_tukeyhsd
    tukey = pairwise_tukeyhsd(analysis_df['Confidence_Score'], analysis_df['Label'])
    print(tukey)
```

```python


#                coclusion:


#      Correct predictions show significantly higher confidence scores (mean ≈ 0.85) compared to incorrect (≈ 0.65) and uncertain labels (≈ 0.55), validated by statistical testing (p < 0.001)
```

```python



#       5   How does the "Noise_Level" affect the "Confidence_Score" of the detections?







import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
df = pd.read_csv('ai_computer_vision_dataset.csv')
print("Summary Statistics:")
print(df[['Noise_Level', 'Confidence_Score']].describe())
pearson_corr = df['Noise_Level'].corr(df['Confidence_Score'], method='pearson')
spearman_corr = df['Noise_Level'].corr(df['Confidence_Score'], method='spearman')
print(f"\nPearson Correlation: {pearson_corr:.2f}")
print(f"Spearman Correlation: {spearman_corr:.2f}")
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Noise_Level', y='Confidence_Score', data=df, 
            scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title("Noise Level vs. Confidence Score (with Regression Line)")
plt.xlabel("Noise Level")
plt.ylabel("Confidence Score")
plt.grid(True)
plt.show()
plt.figure(figsize=(10, 6))
sns.boxplot(x='Noise_Level', y='Confidence_Score', data=df)
plt.title("Confidence Score Distribution by Noise Level")
plt.xlabel("Noise Level")
plt.ylabel("Confidence Score")
plt.xticks(rotation=45)
plt.show()
if len(df['Noise_Level'].unique()) <= 5:  
    groups = [df[df['Noise_Level']==lvl]['Confidence_Score'] for lvl in df['Noise_Level'].unique()]
    f_val, p_val = stats.f_oneway(*groups)
    print(f"\nANOVA Results: F-value = {f_val:.2f}, p-value = {p_val:.4f}")
    if p_val < 0.05:
        print("→ Statistically significant difference between noise levels.")
    else:
        print("→ No significant difference detected.")
```

```python



#       conclusion:



#    The analysis reveals a negative correlation (r = -0.45) between Noise_Level and Confidence_Score, suggesting that higher noise reduces detection confidence. Visualizations confirm this trend, with significant variance in low-noise scenarios. Further preprocessing (e.g., denoising) may improve model reliability."


```

```python



#   7    re translations between specific language pairs (e.g., Chinese → English) less reliable?




import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('ai_computer_vision_dataset.csv')
signs_df = df[df['Category'] == 'Multi-lingual Street Signs'].copy()
def detect_language(text):
    if pd.isna(text):
        return 'Unknown'
    text = str(text)
    if any(char in text for char in ['禁止', '进入']):  
        return 'Chinese'
    elif any(word in text.lower() for word in ['alto', 'cedez', 'le passage']): 
        return 'Spanish/French'
    elif any(word in text.lower() for word in ['yield', 'stop', 'no entry', 'speed limit']):
        return 'English'
    else:
        return 'Other'

signs_df['Target_Language'] = signs_df['Sign_Translation'].apply(detect_language)
signs_df['Language_Pair'] = signs_df['Sign_Language'] + ' → ' + signs_df['Target_Language']
signs_df['Translation_Correct'] = signs_df['Label'].apply(lambda x: 1 if x == 'Correct' else 0)
accuracy_by_pair = signs_df.groupby('Language_Pair')['Translation_Correct'].agg(['mean', 'count'])
accuracy_by_pair = accuracy_by_pair.rename(columns={'mean': 'Accuracy', 'count': 'Sample Size'})
accuracy_by_pair = accuracy_by_pair.sort_values('Accuracy', ascending=False)
plt.tight_layout()
plt.subplots_adjust(top=0.9)  
plt.show()
print("\nTranslation Accuracy by Language Pair:")
print(accuracy_by_pair.round(2))
```

```python

plt.figure(figsize=(12, 7))
ax = sns.barplot(x=accuracy_by_pair.index, y=accuracy_by_pair['Accuracy'], 
                order=accuracy_by_pair.index,
                palette='viridis')
plt.title('Translation Accuracy by Language Pair\n', 
          fontsize=14, fontweight='bold', pad=20)  
for i, (acc, count) in enumerate(zip(accuracy_by_pair['Accuracy'], accuracy_by_pair['Sample Size'])):
    ax.text(i, acc + 0.02, f'{acc:.2f}\n(n={count})', 
            ha='center', va='bottom', fontsize=9)

plt.ylabel('Accuracy Rate', fontsize=12)
plt.xlabel('Language Pair', fontsize=12)
plt.xticks(rotation=45, ha='right') 
plt.ylim(0, 1.1)  
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.subplots_adjust(top=0.9)  
plt.show()

```

```python


#       conclusion 



#      Translation accuracy varies significantly by language pair, with Chinese translations showing the lowest performance. This highlights key areas for model improvement in multilingual sign recognition systems

```

```python


#       8  Is there a relationship between Solution_Steps (numeric value) and prediction accuracy?



import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
math_df = df[df['Category'] == 'Handwritten Math Solutions'].dropna(subset=['Solution_Steps'])
math_df['Correct'] = math_df['Label'].apply(lambda x: 1 if x == 'Correct' else 0)
corr, p_value = pearsonr(math_df['Solution_Steps'], math_df['Correct'])
print(f"Pearson Correlation: {corr:.3f} (p-value: {p_value:.4f})")
plt.show()
```

```python



#  10   Compare the average noise level for "Correct" and "Incorrect" labels across all categories. Is noise level a significant factor in label accuracy? 


import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('ai_computer_vision_dataset.csv')
print("\n=== Image Quality Analysis and Correlation ===")
quality_metrics = ['Blur_Level', 'Brightness_Level', 'Contrast_Level', 'Noise_Level']
for metric in quality_metrics:
    print(f"\nAverage {metric}:")
    print(df.groupby('Category')[metric].mean())
correlation = df[quality_metrics + ['Confidence_Score']].corr()['Confidence_Score'].drop('Confidence_Score')
print("\nCorrelation between image quality and confidence:")
print(correlation)

```

```python


plt.figure(figsize=(8, 6))
colors = ['grey' if x > 0 else 'brown' for x in correlation.values]
correlation.sort_values().plot(kind='barh', color=colors)
plt.title('Correlation of Image Quality Metrics with Confidence Score')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Image Quality Metric')
plt.tight_layout()
plt.show()
```

```python



#      conclusion:



#    The analysis of image quality metrics (Blur_Level, Brightness_Level, Contrast_Level, and Noise_Level) across the dataset categories reveals varying average levels for each metric.

```
